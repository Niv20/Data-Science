### Dynamic Typing

In contrast to C or C++, Python uses dynamic typing. This means that the value itself has a type, but the variable is just a name that points to that value.

This means you can change what value a variable points to, even if the new value is a different type.


### Mutable vs. Immutable

The terms are immutable, which means "cannot be changed," and mutable, which means "can be changed." The basic data types we've seen so far, like integers (int), decimal numbers (float), text (str), and booleans (bool), are all immutable.

Immutable Types (Like Integers and Strings)

This can be confusing. You might ask: "If an integer is immutable, how can I do x = 1 and then x = 2?"

Here is the secret. When you do x = 1, you tell the variable x to point to a "box" in memory that holds the value 1. When you then run x = x + 1, Python first looks at the value x is pointing to (which is 1). Then, it adds 1 + 1 and gets a new value, 2. It creates a brand new box in memory to hold this value 2. Finally, it breaks the old connection and tells x to point to this new box.

You did not change the original box with 1 in it; you just made x point to an entirely new box.

Key Idea: Any time you "change" an immutable value (like an int or str), Python creates a new object in memory and moves the variable's "pointer" to it.

Because of this, these two lines do the exact same thing for immutable types. Both lines create a new object and make x point to it.

## Input and Output

### Input 
To get information from the user, we use the `input()` function.

×‘× ×™×’×•×“ ×œ×©×¤×•×ª ××—×¨×•×ª ×›××• C ×©×‘×”× ×”×™×” ×¦×¨×™×š ×œ×”×“×¤×™×¡ ×”×•×“×¢×” ×‘× ×¤×¨×“, ×‘×¤×™×™×ª×•×Ÿ ××¤×©×¨ ×œ×”×¢×‘×™×¨ ××—×¨×•×–×ª ×›××¨×’×•×× ×˜ ×œÖ¾input() ×›×“×™ ×œ×”×¦×™×’ ×”×•×“×¢×” ×œ××©×ª××© ×œ×¤× ×™ ×©×”×•× ××–×™×Ÿ ××ª ×”×§×œ×˜ ×©×œ×•.

When Python runs `input()`, it pauses the program, shows a blinking cursor, and waits for the user to type something and press **Enter**.

This is the most important rule to remember about input(): No matter what the user types, input() always gives you back a string. So, if you want to work with numbers, you need to convert the string to a number using functions like `int()`.


````python
name = input("What is your name? ")
age = int(input("How old are you? "))

print("Hello, ", name)
print("Next year you will be", age + 1, "years old.")
````

### Output

You can give print() many items at once, separated by commas. It will print them all on one line, with a single space between each item.

You can print variables:

````python
user_name = "Niv"
print("My name is", user_name)
````

You can even do calculations right inside the print() function:

````python
print("My favorite number is", 10*2-3)
````

Here is a bigger example combining text and multiple variables:

````python
price_one_book = 30
price_two_books = 50

print("One book costs", price_one_book, "shekels, and two books cost", price_two_books, "shekels.")
````

by default, the print() function automatically adds an invisible newline character at the end of whatever you asked it to print. It's just like pressing the "Enter" key on your keyboard after the line is finished.
